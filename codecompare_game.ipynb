```python
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
codecompare = pd.read_excel("C:\\Users\\User\\Downloads\\ZYVO\\game\\Prolific_01.xlsx",sheetname='CodeCompare',header=0)
```

### Zero Score Count

```python
codecompare_zero_score = codecompare[codecompare['Score']==0][['ProlificId','Score']]
codecompare_zero_cnt = pd.DataFrame({'zero_cnt': codecompare_zero_score.groupby('ProlificId')['Score'].count()}).reset_index()
fig = codecompare_zero_cnt.plot.bar(x=codecompare_zero_cnt['ProlificId'])
fig = plt.gcf()
```

```python
sns.distplot(codecompare_zero_cnt['zero_cnt'])
fig = plt.gcf()
```

### One Score Accuracy

```python
codecompare_one_score = codecompare[codecompare['Score']==1][['ProlificId','Score']]
codecompare_one_acc = pd.DataFrame({'Accuracy': codecompare_one_score.groupby('ProlificId')['Score'].count()}).reset_index()
codecompare_one_acc['Accuracy'] = codecompare_one_acc['Accuracy'] / 30
sns.distplot(codecompare_one_acc['Accuracy'])
fig = plt.gcf()
```

### Zero Score Average ReactionTime

```python
codecompare_zero_score = codecompare[codecompare['Score']==0][['ProlificId','ReactionTime']]
codecompare_zero_mean = pd.DataFrame({'0_ReactionTime_mean': codecompare_zero_score.groupby('ProlificId')['ReactionTime'].mean()}).reset_index()
fig = codecompare_zero_mean.plot.bar(x=codecompare_zero_mean['ProlificId'])
fig = plt.gcf()
```

### One Score Average ReactionTime

```python
codecompare_one_score = codecompare[codecompare['Score']==1][['ProlificId','ReactionTime']]
codecompare_one_mean = pd.DataFrame({'1_ReactionTime_mean': codecompare_one_score.groupby('ProlificId')['ReactionTime'].mean()}).reset_index()
fig = codecompare_one_mean.plot.bar(x=codecompare_one_mean['ProlificId'])
fig = plt.gcf()
```
