```python
class Element(object):
    def __init__(self,value):
        self.value = value
        self.next = None
        
class LinkedList(object):
    def __init__(self,head=None):
        self.head = head
    
    def append(self,new_element):
        current = self.head
        if current:
            while current.next:
                current = current.next
            current.next = new_element
        else:
            current.head = new_element
        
    def get_element(self,position):
        current = self.head
        counter = 1
        if position < 1:
            return None
        while counter <= position and current:
            if counter == position:
                return current
            current = current.next
            counter += 1
        return None
    
    def insert(self,new_element,position):
        current = self.head
        counter = 1
        if current:
            while counter <= position-1 and current:
                if counter == position-1:
                    new_element.next = current.next
                    current.next = new_element
                current = current.next
                counter+=1
        else:
            current.head = new_element
        return None
        
    def delete(self,value):
        current = self.head
        previous = None
        while current.value != value and current.next:
            previous = current
            current = current.next
        if current.value == value:
            if previous:
                previous.next = current.next
            else:
                self.head = current.next
        return None
    
    def insert_first(self,new_element):
        new_element.next = self.head
        self.head = new_element
        
    def delete_first(self):
        if self.head:
            delete = self.head
            self.head = delete.next
            return delete
        else:
            return None
        
class Stack(object):
    def __init__(self,top=None):
        self.ll = LinkedList(top)
    
    def push(self,new_element):
        self.ll.insert_first(new_element)
        
    def pop(self):
        return self.ll.delete_first()
    
class Queue(object):
    def __init__(self,head=None):
        self.ll = LinkedList(head)
        
    def enqueue(self,new_element):
        self.ll.append(new_element)
        
    def peek(self):
        return self.ll.get_element(1).value
    
    def dequeue(self):
        self.ll.delete_first()
```

```python
e1 = Element(10)
e2 = Element(20)
e3 = Element(30)

ll = LinkedList(e1)
ll.append(e2)
ll.append(e3)
print("List")
print("Position List 1 : " + str(ll.get_element(1).value))
print("Position List 2 : " + str(ll.get_element(2).value))
print("Position List 3 : " + str(ll.get_element(3).value))

print()

e5 = Element(50)
ll.insert(e5,3)
print("Inserted between 2 & 3")
print("Position List 1 : " + str(ll.get_element(1).value))
print("Position List 2 : " + str(ll.get_element(2).value))
print("Position List 3 : " + str(ll.get_element(3).value))
print("Position List 4 : " + str(ll.get_element(4).value))

print()

ll.delete(50)
ll.delete(10)
print("Deleted 50 & 10")
print("Position List 1 : " + str(ll.get_element(1).value))
print("Position List 2 : " + str(ll.get_element(2).value))
```

```python
e1 = Element(10)
e2 = Element(20)
e3 = Element(30)

stack = Stack(e1)

stack.push(e2)
stack.push(e3)
print("Stack")
print("Position Stack 1 : " + str(stack.ll.get_element(1).value))
print("Position Stack 2 : " + str(stack.ll.get_element(2).value))
print("Position Stack 3 : " + str(stack.ll.get_element(3).value))

print()

print('Popped : ' + str(stack.pop().value))
print("Position Stack 1 : " + str(stack.ll.get_element(1).value))
print("Position Stack 2 : " + str(stack.ll.get_element(2).value))
```

```python
e1 = Element(10)
e2 = Element(20)
e3 = Element(30)

que = Queue(e1)
que.enqueue(e2)
que.enqueue(e3)
print("Queue")
print("Position Queue 1 : " + str(que.ll.get_element(1).value))
print("Position Queue 2 : " + str(que.ll.get_element(2).value))
print("Position Queue 3 : " + str(que.ll.get_element(3).value))

print()

print("Peek Queue")
print(que.peek())

print()

print("Dequeue")
que.dequeue()
print("Position Queue 1 : " + str(que.ll.get_element(1).value))
print("Position Queue 2 : " + str(que.ll.get_element(2).value))
```
