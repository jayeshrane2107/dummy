```python
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
switch = pd.read_excel("C:\\Users\\User\\Downloads\\ZYVO\\game\\Prolific_01.xlsx",sheetname='Switch',header=0)
```

### Zero Score Count

```python
switch_zero_score = switch[switch['Score']==0][['ProlificId','Score']]
switch_zero_cnt = pd.DataFrame({'zero_cnt': switch_zero_score.groupby('ProlificId')['Score'].count()}).reset_index()
fig = switch_zero_cnt.plot.bar(x=switch_zero_cnt['ProlificId'])
fig = plt.gcf()
```

```python
sns.distplot(switch_zero_cnt['zero_cnt'])
fig = plt.gcf()
```

### Zero Score Average Time

```python
switch_zero_score = switch[switch['Score']==0][['ProlificId','Time']]
switch_zero_mean = pd.DataFrame({'0_time_mean': switch_zero_score.groupby('ProlificId')['Time'].mean()}).reset_index()
fig = switch_zero_mean.plot.bar(x=switch_zero_mean['ProlificId'])
fig = plt.gcf()
```

```python
sns.distplot(switch[switch['Score']==0]['Time'])
fig = plt.gcf()
```

### One Score Average Time

```python
switch_one_score = switch[switch['Score']==1][['ProlificId','Time']]
switch_one_mean = pd.DataFrame({'1_time_mean': switch_one_score.groupby('ProlificId')['Time'].mean()}).reset_index()
fig = switch_one_mean.plot.bar(x=switch_one_mean['ProlificId'])
fig = plt.gcf()
```

```python
sns.distplot(switch[switch['Score']==1]['Time'])
fig = plt.gcf()
```

### Time Distribution

```python
sns.distplot(switch['Time'])
fig = plt.gcf()
```
